### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici
# features_by_species a la shape (3, 50, 4)
variance_by_species = np.var(features_by_species, axis=1) # Variance des 4 features pour chaque espèce
total_variance_by_species = np.sum(variance_by_species, axis=1) # Somme des variances des 4 features

species_labels = ['Setosa', 'Versicolor', 'Virginica']
max_variance_index = np.argmax(total_variance_by_species)
most_variable_species = species_labels[max_variance_index]

print(f"Variances totales par espèce: {total_variance_by_species}")
print(f"L'espèce avec la plus grande variance globale est: {most_variable_species}")

# Visualisation des distributions de l'espèce la plus variable (Virginica, index 2)
# Nous allons utiliser un boxplot par feature pour cette espèce
virginica_features = features_by_species[max_variance_index]

plt.figure(figsize=(8, 6))
plt.boxplot(virginica_features, labels=feature_names)
plt.title(f'Distribution des Caractéristiques pour {most_variable_species} (Espèce la Plus Variable)')
plt.ylabel('Longueur/Largeur (cm)')
plt.grid(axis='y', alpha=0.5)
plt.show()
```
